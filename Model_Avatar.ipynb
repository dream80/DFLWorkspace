# 第四步. 训练模型

# 第五步. 转换输出

```
#用src中的脸替换dst的脸 
!python main.py convert --input-dir ../workspace/data_dst --output-dir ../workspace/data_dst/merged --aligned-dir ../workspace/data_dst/aligned --model-dir ../workspace/model --model AVATAR
```

```
#把替换好的图片转换成视频

!python main.py videoed video-from-sequence --input-dir ../workspace/data_dst/merged --output-file ../workspace/result.mp4 --reference-file ../workspace/data_dst.mp4
```
